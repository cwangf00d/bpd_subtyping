```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_DIR_PATH = '/Users/cindywang/PycharmProjects/bpd-subtyping/data/'
```

```python
tar_df = pd.read_csv(DATA_DIR_PATH + 'processed/full/tar_df.csv')
dis_df = pd.read_csv(DATA_DIR_PATH + 'discharge_bpd_status.csv')
tardis_df = tar_df.merge(dis_df.loc[:, ['PatientSeqID', 'BPD Grade']])
```

```python
tardis_df.head()
```

```python
# now, something I'm interested in looking at is 
# how the probabilities predicted for risk change for the grade 3 cohort + the premature death cohort
plt.figure()
sns.lineplot(data=tardis_df, x='DSB', y='grade_minus1_prob', hue='BPD Grade')
plt.title('Trajectory of Risk Probabilities for Premature Death Outcome Given Patients of Varying BPD Grades')
plt.ylabel('Probability of Risk for Premature Death')

plt.figure()
sns.lineplot(data=tardis_df, x='DSB', y='grade_3_prob', hue='BPD Grade')
plt.title('Trajectory of Risk Probabilities for Grade 3 BPD Given Patients of Varying BPD Grades')
plt.ylabel('Probability of Risk for Grade 3 BPD')

plt.figure()
sns.lineplot(data=tardis_df, x='DSB', y='grade_2_prob', hue='BPD Grade')
plt.title('Trajectory of Risk Probabilities for Grade 2 BPD Given Patients of Varying BPD Grades')
plt.ylabel('Probability of Risk for Grade 2 BPD')

plt.figure()
sns.lineplot(data=tardis_df, x='DSB', y='grade_1_prob', hue='BPD Grade')
plt.title('Trajectory of Risk Probabilities for Grade 1 BPD Given Patients of Varying BPD Grades')
plt.ylabel('Probability of Risk for Grade 1 BPD')

plt.figure()
sns.lineplot(data=tardis_df, x='DSB', y='grade_0_prob', hue='BPD Grade')
plt.title('Trajectory of Risk Probabilities for No BPD Given Patients of Varying BPD Grades')
plt.ylabel('Probability of Risk for No BPD')
```
